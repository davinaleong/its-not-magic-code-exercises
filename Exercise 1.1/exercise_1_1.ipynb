## 2. Read the Folder's Contents

Before renaming anything, list out what's actually in your folder. Set folder to wherever your reports live, then loop through it and print each filename.

```python
folder = Path("exercise")

# TODO: loop through the folder and print each file's name
# (skip subfolders — we only want files)
for file in folder.iterdir():
    print(file.name)
```

## 3. Generate a Date Format in YYYYMMDD

Generate today's date as a string in `YYYYMMDD` format (e.g. `20260702`). This format sorts correctly when filenames are viewed alphabetically — that's why it's a popular convention.

```python
# TODO: create a variable called date_str holding today's date as YYYYMMDD
date_str = datetime.now().strftime("%Y%m%d")
print(date_str)
```

## 4. Preview File Renaming

Never rename blindly — always preview first. Loop through the folder again, but this time print what each new filename would look like (old name → new name), without actually renaming anything yet.

```python
# TODO: loop through the folder and print "old_name -> new_name"
# for each file, using date_str as a prefix
for file in folder.iterdir():
    if file.is_file():
            new_name = f"{date_str}_{file.name}"
            print(f"{file.name} -> {new_name}")
```

## 5. Rename Files in Folder

Now that you've previewed and it looks correct, do the real rename.

```python
# TODO: loop through the folder and actually rename each file
# using the same date-prefix logic as step 4
for file in folder.iterdir():
    if file.is_file():
        new_name = f"{date_str}_{file.name}"
        file.rename(folder / new_name)

print("Done renaming.")
```

### 1. Use each file's last modified date instead of today's date.

```python
folder = Path("challenge_1")

for file in folder.iterdir():
    if file.is_file():
        mod_date = datetime.fromtimestamp(file.stat().st_mtime)
        date_str = mod_date.strftime("%Y%m%d")
        new_name = f"{date_str}_{file.name}"
        file.rename(folder / new_name)
        print(f"Renamed: {file.name} -> {new_name}")

print("Done renaming.")
```
