# Exercise 1.1

Auto-rename a folder of files. Take 50 downloaded reports and rename them with a date prefix so they finally stop looking like “final_final_v7_reallyfinal.pdf”. You will practise strings, loops, and file handling.

## 1. Imports

You'll need two things from Python's standard library for this exercise — no installs required.
💭 Think about it: What do you need to (a) work with files/folders, and (b) work with today's date?

```python
import os
from pathlib import Path
from datetime import datetime
import re
```

## 2. Read the Folder's Contents

Before renaming anything, list out what's actually in your folder. Set folder to wherever your reports live, then loop through it and print each filename.

```python
folder = Path("exercise")

for item in folder.iterdir():
    print(item.name)
```

## 3. Generate a Date Format in YYYYMMDD

Generate today's date as a string in `YYYYMMDD` format (e.g. `20260702`). This format sorts correctly when filenames are viewed alphabetically — that's why it's a popular convention.

```python
date_str = datetime.now().strftime("%Y%m%d")
print(date_str)
```

## 4. Preview File Renaming

Never rename blindly — always preview first. Loop through the folder again, but this time print what each new filename would look like (old name → new name), without actually renaming anything yet.

```python
for file in folder.iterdir():
    if file.is_file():
        new_name = f"{date_str}_{file.name}"
        print(f"{file.name}  ->  {new_name}")   # preview only, no rename yet
```

## 5. Rename Files in Folder

Now that you've previewed and it looks correct, do the real rename.

```python
for file in folder.iterdir():
    if file.is_file():
        new_name = f"{date_str}_{file.name}"
        file.rename(folder / new_name)

print("Done renaming.")
```

## 6. Challenge (optional)

Once your basic version works, try leveling it up:

### 1. Use each file's last modified date instead of today's date.

```python
folder = Path("challenge_1")

for file in folder.iterdir():
    if file.is_file():
        mod_time = datetime.fromtimestamp(file.stat().st_mtime)
        date_str = mod_time.strftime("%Y%m%d")
        new_name = f"{date_str}_{file.name}"
        file.rename(folder / new_name)
        print(f"Renamed: {file.name}  ->  {new_name}")

print("Done renaming.")
```

### 2. Skip files that already have a date prefix, so re-running the notebook doesn't double them up.

```python
folder = Path("challenge_2")

already_prefixed = re.compile(r"^\d{8}_")

for file in folder.iterdir():
    if file.is_file():
        if already_prefixed.match(file.name):
            print(f"Skipping (already prefixed): {file.name}")
            continue
        new_name = f"{date_str}_{file.name}"
        file.rename(folder / new_name)
        print(f"Renamed: {file.name}  ->  {new_name}")

print("Done renaming.")
```

### Only rename files with a specific extension, like .pdf.

```python
folder = Path("challenge_3")

target_extension = ".pdf"

for file in folder.iterdir():
    if file.is_file() and file.suffix.lower() == target_extension:
        new_name = f"{date_str}_{file.name}"
        file.rename(folder / new_name)
        print(f"Renamed: {file.name}  ->  {new_name}")

print("Done renaming.")
```
