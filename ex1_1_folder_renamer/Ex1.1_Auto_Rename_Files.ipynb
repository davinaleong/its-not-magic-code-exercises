## The scenario
You've been handed a folder of 50 downloaded reports. The names are chaos. Your job: rename every file with a clean `YYYY-MM-DD_` prefix based on when it was last modified, so they sort chronologically and finally make sense.

**A sample messy folder (`messy_reports/`) has been generated for you** — 50 real dummy files with genuinely terrible names, sitting right next to this notebook. Let's fix them.

## Step 1: Look before you leap

Before writing any code that *changes* files, always look at what you're working with first. This one habit will save you from a lot of regret.

````python
import os

folder = "messy_reports"

# os.listdir() gives us every filename in the folder, as a list of strings
all_files = os.listdir(folder)

print(f"Found {len(all_files)} files.")
print("First 10 filenames:")
for filename in all_files[:10]:
    print(" -", filename)
````

**What just happened?**

- `os.listdir(folder)` returns a **list** — think of it as Python's version of a to-do list, where each item is a filename (a string).
- The `for filename in all_files[:10]:` line is a **loop**. It says: "for each item in this list, do the following thing." The `[:10]` just means "only the first 10," so we don't flood the screen.

You'll use this exact pattern — list the files, loop through them — in almost every file-handling script you ever write.

## Step 2: Get the date each file was last modified

We want to prefix each file with the date it was last changed, e.g. `2025-03-14_Budget_final.xlsx`. Python can ask the operating system for this — we don't have to guess or open the file.

````python
import datetime

# Let's inspect one file first
sample_file = all_files[0]
sample_path = os.path.join(folder, sample_file)

# os.path.getmtime() returns the "last modified" time as a number
# (seconds since 1 Jan 1970 -- a computer's favourite starting point)
timestamp = os.path.getmtime(sample_path)

# datetime.fromtimestamp() turns that number into something human-readable
modified_date = datetime.datetime.fromtimestamp(timestamp)

print("File:", sample_file)
print("Raw timestamp:", timestamp)
print("Human-readable:", modified_date)
print("Just the date, formatted:", modified_date.strftime("%Y-%m-%d"))
````

**Key idea:** `strftime` (string-format-time) is how Python turns a date into text laid out exactly the way you want. `%Y-%m-%d` means "4-digit year, 2-digit month, 2-digit day" — which is also the format that sorts correctly alphabetically. That's not an accident; it's why professionals use it.

## Step 3: Build the new filename

Now we combine the date with the original filename. `os.path.splitext()` splits a filename into the name and the extension (`.pdf`, `.docx`, etc.) so we can insert our prefix *before* the extension, not after it.

````python
name_only, extension = os.path.splitext(sample_file)
print("Name:", name_only)
print("Extension:", extension)

date_prefix = modified_date.strftime("%Y-%m-%d")
new_name = f"{date_prefix}_{name_only}{extension}"

print("Old name:", sample_file)
print("New name:", new_name)
````

## Step 4: Do it for ALL 50 files (dry run first!)

Here's the golden rule of any script that changes or deletes files: **do a dry run first.** Print what *would* happen before you actually let it happen. It costs you nothing and saves you everything.

````python
for filename in all_files:
    filepath = os.path.join(folder, filename)

    # skip anything that isn't actually a file (just in case)
    if not os.path.isfile(filepath):
        continue

    timestamp = os.path.getmtime(filepath)
    modified_date = datetime.datetime.fromtimestamp(timestamp)
    date_prefix = modified_date.strftime("%Y-%m-%d")

    name_only, extension = os.path.splitext(filename)
    new_name = f"{date_prefix}_{name_only}{extension}"

    print(f"{filename}  -->  {new_name}")
````

Scan through the output above. Does it look right? Are there any filenames that look odd or would collide with another? Good — now let's actually rename them.

````python
renamed_count = 0

for filename in all_files:
    filepath = os.path.join(folder, filename)

    if not os.path.isfile(filepath):
        continue

    timestamp = os.path.getmtime(filepath)
    modified_date = datetime.datetime.fromtimestamp(timestamp)
    date_prefix = modified_date.strftime("%Y-%m-%d")

    name_only, extension = os.path.splitext(filename)
    new_name = f"{date_prefix}_{name_only}{extension}"
    new_path = os.path.join(folder, new_name)

    os.rename(filepath, new_path)
    renamed_count += 1

print(f"Done! Renamed {renamed_count} files.")
````

````python
# Let's verify - list the folder again, sorted this time
renamed_files = sorted(os.listdir(folder))
for f in renamed_files[:10]:
    print(f)
print("...")
print(f"Total files: {len(renamed_files)}")
````

## 🎉 You did it

Your 50 chaotic filenames are now sorted, dated, and readable at a glance. Notice what you actually learned to *see*:
- A folder is just a list you can loop over
- A filename is just a string you can slice and rebuild
- "Automate this" usually means "loop + small transformation, repeated"

That pattern — **loop through items, do a small transformation, repeat** — is at the heart of an enormous amount of everyday automation, whether you write it yourself or ask an AI to write it for you.
