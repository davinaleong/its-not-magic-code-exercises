## The scenario
`recipients.csv` has 12 names and departments. We'll generate a personalised subject line for each, ready to paste into a mail-merge tool.

## Step 1: Load the recipient list

```python
import csv

with open("recipients.csv", newline="") as f:
    reader = csv.DictReader(f)
    recipients = list(reader)

print(f"Loaded {len(recipients)} recipients")
for r in recipients[:3]:
    print(r)
```

## Step 2: The simplest possible personalisation

An f-string (`f"..."`) lets you drop a variable straight into a piece of text using `{curly_braces}`. This is the foundation of everything else in this exercise.

```python
name = recipients[0]["name"]
first_name = name.split()[0]   # just take the first word of the name

subject = f"{first_name}, your Q3 report is ready"
print(subject)
```

**Why `name.split()[0]`?** `"Alex Tan".split()` breaks the string into `["Alex", "Tan"]` at the space, and `[0]` grabs the first item. It's more natural to write "Hi Alex" than "Hi Alex Tan" in a subject line — small touch, but it's the kind of detail that makes automated messages feel less robotic.

## Step 3: Do it for everyone, with a loop

```python
subject_lines = []

for r in recipients:
    first_name = r["name"].split()[0]
    department = r["department"]
    subject = f"{first_name}, your {department} team's Q3 report is ready"
    subject_lines.append({"name": r["name"], "email_subject": subject})

for item in subject_lines:
    print(item["email_subject"])
```

## Step 4: Vary the template by department (a small taste of conditionals)

```python
def make_subject(name, department):
    first_name = name.split()[0]

    if department == "Sales":
        return f"{first_name}, your pipeline numbers just landed 📈"
    elif department == "Finance":
        return f"{first_name}, this month's figures are in - please review"
    elif department == "Engineering":
        return f"{first_name}, sprint report ready for your eyes"
    else:
        return f"{first_name}, your {department} team's Q3 report is ready"

for r in recipients:
    print(make_subject(r["name"], r["department"]))
```

**What just changed?** We wrapped the logic in a **function** — `def make_subject(name, department):` — so we can reuse it cleanly, and added an `if / elif / else` chain so different departments get a slightly different tone. This is the same core idea as Step 3, just with a bit more nuance per recipient.

## Step 5: Save the results for a mail merge

```python
with open("subject_lines_output.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "email_subject"])
    writer.writeheader()
    for r in recipients:
        writer.writerow({"name": r["name"], "email_subject": make_subject(r["name"], r["department"])})

print("Saved subject_lines_output.csv - ready to plug into your mail merge tool.")
```
